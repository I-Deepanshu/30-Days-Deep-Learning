```python
from PIL import Image
from IPython.display import display
import os
import torch 

# Load YOLOv5 pre-trained model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Replace this with the path to your local image
img = '/Users/raghavgarg/Downloads/archive (1)/seg_pred/seg_pred/3.jpg'  # Absolute path to your image

# Run the YOLOv5 model
results = model(img)
results.save()  # Save predictions locally

# Dynamically find the latest 'exp' folder in runs/detect
save_dir = 'runs/detect'
latest_exp = sorted(os.listdir(save_dir))[-1]  # Most recent 'exp' folder
output_path = os.path.join(save_dir, latest_exp, os.listdir(os.path.join(save_dir, latest_exp))[0])

# Print the output path
print("Output image saved to:", output_path)

# Display the processed image
display(Image.open(output_path))
```
