```python
import tensorflow as tf  
from tensorflow.keras import Sequential  
from tensorflow.keras.layers import Dense  

# Sample Data  
x = tf.random.normal((500, 10))  
y = tf.random.normal((500, 1))  

# Define Model  
model = Sequential([  
    Dense(64, activation='relu'),  
    Dense(32, activation='relu'),  
    Dense(1)  
])  

model.compile(optimizer='adam', loss='mse')  
model.fit(x, y, epochs=10, batch_size=32)  
```

```python
import torch  
import torch.nn as nn  
import torch.optim as optim  

# Sample Data  
x = torch.randn(500, 10)  
y = torch.randn(500, 1)  

# Define Model  
class SimpleNN(nn.Module):  
    def __init__(self):  
        super(SimpleNN, self).__init__()  
        self.net = nn.Sequential(  
            nn.Linear(10, 64),  
            nn.ReLU(),  
            nn.Linear(64, 32),  
            nn.ReLU(),  
            nn.Linear(32, 1)  
        )  

    def forward(self, x):  
        return self.net(x)  

model = SimpleNN()  
criterion = nn.MSELoss()  
optimizer = optim.Adam(model.parameters(), lr=0.001)  

# Training  
for epoch in range(10):  
    optimizer.zero_grad()  
    outputs = model(x)  
    loss = criterion(outputs, y)  
    loss.backward()  
    optimizer.step()  
    print(f"Epoch {epoch+1}, Loss: {loss.item()}")  
```

```python
import matplotlib.pyplot as plt  

frameworks = ['TensorFlow', 'PyTorch']  
times = [18.3, 16.7]  
memory = [490, 470]  

fig, ax1 = plt.subplots()  

# Training Time  
ax1.bar(frameworks, times, color='lightblue', label='Training Time')  
ax1.set_ylabel('Training Time (s)', color='blue')  

# GPU Memory  
ax2 = ax1.twinx()  
ax2.plot(frameworks, memory, color='orange', marker='o', label='GPU Memory')  
ax2.set_ylabel('GPU Memory (MB)', color='orange')  

plt.title('Framework Comparison: TensorFlow vs PyTorch')  
plt.show()  
```
